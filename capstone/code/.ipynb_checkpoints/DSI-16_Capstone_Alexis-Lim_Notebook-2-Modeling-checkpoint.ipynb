### Model Functions

```python
# Define functions to track our model metrics

# Function to return R^2 adjusted
def r2_adj(y_true, y_preds, p):
    n = len(y_true)
    y_mean = np.mean(y_true)
    numerator = np.sum(np.square(y_true - y_preds)) / (n - p - 1)
    denominator = np.sum(np.square(y_true - y_mean)) / (n - 1)
    return 1 - (numerator / denominator)

# Function to print metrics; returns R^2, adjusted R^2 and RMSE
def get_metrics(y_true, y_pred, p):
    
    mse = metrics.mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = metrics.median_absolute_error(y_true, y_pred)
    r2 = metrics.r2_score(y_true, y_pred)
    r2a = r2_adj(y_true, y_pred, p)
    
    print('Mean squared error      = ', mse )
    print('Root mean squared error = ', rmse)
    print('Mean absolute error   = ', mae)
    print('R^2                     = ', r2)
    print('Adjusted R^2            = ', r2a)
    return round(r2, 5), round(r2a, 5), round(rmse, 5)

# Function to get model metrics
def model_metrics(model, mname, X_train_sc, y_train, X_test_sc, y_test):
    
    y_pred_train = model.predict(X_train_sc)
    y_pred = model.predict(X_test_sc)
    
    p = X_train_sc.shape[1]
    
    print(f'{mname} Metrics:')
    print("*" * 25)
    print('Performance on Train Data')
    print("-" * 25)
    trainr2, trainr2a, trainrmse = get_metrics(y_train, y_pred_train, p)
    print('')
    print('Performance on Test Data')
    print("-" * 25)
    testr2, testr2a, testrmse = get_metrics(y_test, y_pred, p)
    
    return [mname, p, trainr2, trainr2a, trainrmse, testr2, testr2a, testrmse]
```

```python
# Define a dataframe to keep track of our metrics
model_tracker = pd.DataFrame(columns=['Method','Variable Count','Train R^2 Score','Train Adj R^2 Score',
                                      'Train RMSE','Test R^2 Score','Test Adj R^2 Score','Test RMSE'])
```

## One Hot Encoding

Our categorical columns need to be one-hot encoded into additional columns. We will do this for the product category, as well as on the states and zip codes. For the latter two, we will only choose to use one set in the final model, but we will compare the performance of both in our modeling.

```python
# Create one hot encoding for product categories
product_dummy = pd.get_dummies(sellers['most_common_cat'])

# Merge dummies with sellers df
sellers = pd.merge(sellers, product_dummy, left_index=True, right_index=True)
```

```python
# Create one hot encoding for states
state_dummy = pd.get_dummies(sellers['state'])

# Create list to store states
states = state_dummy.columns.to_list()

# Merge dummies with sellers df
sellers = pd.merge(sellers, state_dummy, left_index=True, right_index=True)
```

```python
# Create one hot encoding for zip codes
zip_dummy = pd.get_dummies(sellers['zip_code_prefix_2_digits'])

# Create list to store zip codes
zip_codes = zip_dummy.columns.tolist()

# Merge dummies with sellers df
sellers = pd.merge(sellers, zip_dummy, left_index=True, right_index=True)
```

Since the performance of zip codes versus states is relatively similar, we can decide to go with the states category instead to reduce the complexity of our model. 

Given that the model appears to be facing challenges in predicting values with higher values and we are getting negative values for our Gradient Booster models, we'll take a look at modeling with the average sales per month as our target variable.

However, we will need to take into account the log function when we calculate our evaluation metrics.

```python
# Define columns to drop for log model
columns_to_drop = ['seller_id','most_common_cat','state','zip_code_prefix_2_digits', 
                          'average_sales_per_month','average_sales_per_month_log'] + zip_codes
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
ss = StandardScaler()

X_train_scaled = ss.fit_transform(X_train)
X_test_scaled = ss.transform(X_test)
```

```python
# Function to get metrics (adjusted for log y)
def get_metrics_log(y_true, y_pred, p):
    
    y_true = np.exp(y_true)
    y_pred = np.exp(y_pred)
    
    mse = metrics.mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = metrics.median_absolute_error(y_true, y_pred)
    r2 = metrics.r2_score(y_true, y_pred)
    r2a = r2_adj(y_true, y_pred, p)
    
    print('Mean squared error      = ', mse)
    print('Root mean squared error = ', rmse)
    print('Mean absolute error   = ', mae)
    print('R^2                     = ', r2)
    print('Adjusted R^2            = ', r2a)
    return round(r2, 5), round(r2a, 5), round(rmse, 5)

# Function to get model metrics (adjusted log)
def model_metrics_log(model, mname, X_train_sc, y_train, X_test_sc, y_test):
    
    y_pred_train = model.predict(X_train_sc)
    y_pred = model.predict(X_test_sc)
    
    p = X_train_sc.shape[1]
    
    print(f'{mname} Metrics:')
    print("*" * 25)
    print('Performance on Train Data')
    print("-" * 25)
    trainr2, trainr2a, trainrmse = get_metrics_log(y_train, y_pred_train, p)
    print('')
    print('Performance on Test Data')
    print("-" * 25)
    testr2, testr2a, testrmse = get_metrics_log(y_test, y_pred, p)
    return [mname, p, trainr2, trainr2a, trainrmse, testr2, testr2a, testrmse]
```

# Reviews Topic Modeling

For our review data, we want to see if we can identify clear topics from our positive, neutral and negative reviews and use these topics to generate insights for our sellers and team.

Since we have a scale of 1-5, we classified the reviews into the following categories:
- Positive: 5
- Neutral: 3-4
- Negative: 1-2

```python
reviews = pd.read_csv("../datasets/processed/review_modeling.csv")
```

```python
reviews.head()
```

```python
# Tokenize our string to help create our dictionary for the LDA model

def tokenize(string):
    return string.split()

reviews['tokenized'] = reviews['cleaned_text'].apply(tokenize)
```

```python
reviews_positive = reviews[reviews['review_score'] == 5]
reviews_neutral = reviews[(reviews['review_score'] < 5) & (reviews['review_score'] > 2)]
reviews_negative = reviews[reviews['review_score'] <= 2]
```

```python
print(reviews_positive.shape)
print(reviews_neutral.shape)
print(reviews_negative.shape)
```

## Positive Reviews

To perform our topic modeling, we will rely on Latent Dirichlet Allocation, one of the most popular algorithms for topic modeling. The LDA model requires the id2word, which is a dictionary that maps a unique ID to each word, as well as the corpus, which, for every document, indicates the number of each word that is present based on the dictionary.

```python
# Creating our dictionary and filtering out words that do not appear more than 3 times.
dictionary_pos = Dictionary(reviews_positive['tokenized'])
dictionary_pos.filter_extremes(no_below=3, keep_n=None)

# Assign new ids for words based on filtered list
dictionary_pos.compactify()
```

```python
# Create the corpus
corpus_pos = [dictionary_pos.doc2bow(doc) for doc in reviews_positive['tokenized']]
```

**Number of topics**

We'll try a range of topics to see which number of topics gives us intepretable results that makes sense.

```python
lda_pos = models.LdaModel(corpus=corpus_pos, id2word=dictionary_pos, 
                          num_topics=2, chunksize = 500, random_state=42, passes=10)
lda_pos.print_topics()
```

```python
lda_pos_3 = models.LdaModel(corpus=corpus_pos, id2word=dictionary_pos, 
                          num_topics=3, chunksize = 500, random_state=42, passes=10)
lda_pos_3.print_topics()
```

```python
lda_pos_4 = models.LdaModel(corpus=corpus_pos, id2word=dictionary_pos, 
                          num_topics=4, chunksize = 500, random_state=42, passes=10)
lda_pos_4.print_topics()
```

```python
lda_pos_5 = models.LdaModel(corpus=corpus_pos, id2word=dictionary_pos, 
                          num_topics=5, chunksize = 500, random_state=42, passes=10)
lda_pos_5.print_topics()
```

```python
lda_pos_6 = models.LdaModel(corpus=corpus_pos, id2word=dictionary_pos, 
                          num_topics=6, chunksize = 500, random_state=42, passes=10)
lda_pos_6.print_topics()
```

Based on the topics that we see above, the most salient number of topics seems to be 4 topics. We can visualize this using pyldaviz to get a sense of how the topics may be related.

```python
vis_pos = pyLDAvis.gensim.prepare(lda_pos_4, corpus_pos, dictionary_pos)
```

```python
pyLDAvis.display(vis_pos)
```

These four topics seem to correspond to:
1. Satisfaction with product and service, and future recommendation
2. Satisfaction with the fast delivery and condition of the product
3. Satisfaction with the store and future recommendation
4. Satisfaction with the delivery and product

Topics 1 and 3 seem more closely related and topics 2 and 4 seem more closely related. Overall the happiest customers talk about the quality of the product and the delivery times. We would also recommend an actication like sending these customers a referral code for their friends since they are willing to recommend the brand and product.

## Data Limitations

**Sales Data**

As this is a sample of the data, more accurate observations and predictions can be obtained if we are able to access a more comprehensive set of data, for example all of the data for selected sellers across a 6 month period. This will also allow us to make time-based predictions by splitting the data across months and improve the model.

**Reviews Data**

With a larger corpus of data, we may be able to identify more salient topics, as the ones we have managed to find here give us an idea of the potential of topic modeling for reviews.

As this is a sample of data taken on orders where we have review data, it may be fruitful for us as well to investigate our orders where we have no reviews, and why, so we can identify ways to get customers to give us more feedback.

## Further Investigation/Extensions

With more time and more data, it would be interesting to explore:

**Predicting Sales Volume**

Flipping the problem around, we can try to predict the optimal sales volume for customers to sell their product at for both Olist and the seller to maximise our returns.

**Additional Factors**

If we have access to more data, we can incorporate additional information such as seasonal variables, seller relationship length, additional seller characteristics (size of company, frequency of updates of catalog etc.).

We can also further refine our data by clustering our locations and categories.

**Customer Lifetime Value**

Due to a lack of sufficient repeat customer data, we could not examine the data from a customer point of view, but it will also be interesting to look at CLV, especially if we can consider the platforms and marketplaces on which the customers are buying from us.
